```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargamos el dataset
df = pd.read_csv('../data/Teen_Mental_Health_Dataset.csv')
df.head()
```

```python
# Ver tipos de datos y si hay nulos
print(df.info())

# Estadísticas descriptivas
df.describe()
```

```python
# ¿Cómo afecta el tiempo en redes sociales al nivel de ansiedad?
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='daily_social_media_hours', y='anxiety_level', hue='gender')
plt.title('Relación entre Horas en Redes y Ansiedad')
plt.show()
```
